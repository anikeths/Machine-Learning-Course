# OneHotEncoder

<font color='red'>Assignment:</font> **Neighborhood** is a categorical feature. It has to be transformed before feeding it to a machine learning model. Please transform the column in train_raw to a data frame X_train using **OneHotEncoder**. 

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
# creating instance of one-hot-encoder
one_hot_cats = train_raw.Neighborhood.unique()
one_hot_cats = list(one_hot_cats)
enc = OneHotEncoder(categories = [one_hot_cats])

#perform one-hot encoding on 'Neighborhood' column 
#print(train_raw["Neighborhood"].shape)
enc_df = pd.DataFrame(enc.fit_transform(train_raw[['Neighborhood']]).toarray())

col_to_drop=['Neighborhood']
#merge one-hot encoded columns back with original DataFrame
#X_train= train_raw.drop(columns=col_to_drop).join(enc_df)

df1 = train_raw.drop(columns=col_to_drop).reset_index()
df2 = enc_df.reset_index()
X_train = pd.concat([df1, df2], axis=1)


print(train_raw.drop(columns=col_to_drop).shape)
print(enc_df.shape)
print(X_train.shape)
print(enc_df)

#view final df
X_train=X_train.drop(['index','index'],axis=1)
print(X_train)
```

<font color='red'>Question:</font> What does OneHotEncoder do? Why do we set handle_unknown='ignore'? What is a sparse matrix (try set sparse=True and repeat to see what a sparse matrix looks like).

### One hot encoding can be defined as the essential process of converting the categorical data variables to be provided to machine and deep learning algorithms which in turn improve predictions as well as classification accuracy of a model. One Hot Encoding is a common way of preprocessing categorical features for machine learning models.

### Sparse Matrix: In numerical analysis and scientific computing, a sparse matrix or sparse array is a matrix in which most of the elements are zero.


### Setting handle_unknown = 'error' or 'ignore' - It can be set to either of these.

### handle_unknown{‘error’, ‘ignore’}, default=’error’ - it is set to 'error' by default.
### Whether to raise an error or ignore if an unknown categorical feature is present during transform (default is to raise).

### Rationale behind setting it to 'ignore':

### When this parameter is set to ‘ignore’ and an unknown category is encountered during transform, the resulting one-hot encoded columns for this feature will be all zeros. In the inverse transform, an unknown category will be denoted as None.

### Set sparse matrix as False

```python
enc = OneHotEncoder(handle_unknown='ignore',sparse=False)
enc.fit(df)
OneHotEncoder(handle_unknown='ignore')
enc.categories_
```

### Set sparse matrix as true.

```python
enc = OneHotEncoder(handle_unknown='ignore',sparse=True)
#>>> X = [['Male', 1], ['Female', 3],['Female', 2]]
enc.fit(df)
OneHotEncoder(handle_unknown='ignore')
print(enc.categories_)
```

### Setting handle_unknown as Error

```python
enc = OneHotEncoder(handle_unknown='error',sparse=True)
#>>> X = [['Male', 1], ['Female', 3],['Female', 2]]
enc.fit(df)
OneHotEncoder(handle_unknown='ignore')
print(enc.categories_)
```
